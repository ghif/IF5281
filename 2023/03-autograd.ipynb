```python
import torch
```

### Differentiable Tensor (Enabling requires_grad)

```python
# Create a 2x2 matrix
W = torch.tensor([[1,2], [3, 4]], requires_grad=True, dtype=torch.float)
W
```

```python
# Do an operation
Y = W - 2
Y
```

```python
Y.grad_fn
```

```python
Y = Y * Y + 3
a = Y.mean()
a
```

```python
from torchviz import make_dot
```

```python
make_dot(a)
```

### Gradients

```python
a.backward()
```

Print $\frac{\delta a}{\delta W}$

```python
W.grad
```

### Backprop

```python
W = W - 0.0001 * W.grad
W
```

```python
n = 3
x = torch.arange(1., n+1)
w = torch.ones(n, requires_grad=True)
```

```python
y = w @ x
y
```

```python
y.backward()
```

```python
w.grad
```

```python
x.grad
```
